#### 3. Provide two real-world scenarios where removing stopwords can be harmful.
Removing stopwords can sometimes harm the meaning of text, especially in tasks where context and intent are important. One real-world scenario is sentiment analysis, where words like “not” or “no” are crucial. For example, the sentence “I am not happy” would become “I happy” if stopwords are removed, completely reversing the sentiment from negative to positive. Another scenario is in question-answering or search systems, where stopwords help define the intent of a query. For instance, in the question “What is artificial intelligence?”, removing words like “what” and “is” would leave “artificial intelligence”, which loses the interrogative intent and may lead to incorrect or incomplete responses.


#### 4. Explain the difference between stemming and lemmatization.
Stemming and lemmatization are both techniques used in Natural Language Processing (NLP) to reduce words to their base or root form, but they differ in approach and accuracy. Stemming is a rule-based method that simply removes prefixes or suffixes from words without considering their meaning, which often results in incomplete or non-dictionary words; for example, “studies” may become “studi.” It is fast and computationally efficient but less accurate. In contrast, lemmatization is a more advanced, dictionary-based approach that considers the context and grammatical structure of a word (such as its part of speech) to convert it into a meaningful base form, known as a lemma. For instance, “studies” becomes “study” and “better” can become “good.” While lemmatization is slower than stemming, it produces more accurate and meaningful results, making it more suitable for tasks where understanding language context is important.

## Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    # Handle invalid input
    if text is None or not isinstance(text, str):
        return []

    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (≤2), keep 'no', 'not'
    filtered_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ['no', 'not']:
            filtered_tokens.append(word)

    return filtered_tokens


print(preprocess_text("I absolutely looooved this product 😍😍"))
```

## Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    clean_sentence = " ".join(tokens)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean_sentence)
    print("-" * 50)

    results.append(tokens)
```

## Task 4: Token Analytics

```python
import numpy as np

analytics = []

for tokens in results:
    if len(tokens) == 0:
        total = unique = avg_len = 0
    else:
        total = len(tokens)
        unique = len(set(tokens))
        avg_len = np.mean([len(t) for t in tokens])
    analytics.append((total, unique, avg_len))

# Display analytics
for i, (t, u, a) in enumerate(analytics):
    print(f"Sentence {i+1}: Total={t}, Unique={u}, Avg Length={a:.2f}")
```

## Task 5: Frequency Analysis

```python
from collections import Counter
# Combine all tokens
all_tokens = [token for sublist in results for token in sublist]
counter = Counter(all_tokens)
print("Top 10 Most Frequent Words:")
print(counter.most_common(10))
print("\nTop 5 Least Frequent Words:")
least_common = counter.most_common()[:-6:-1]
print(least_common)
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
test = [
    "Hello!!! 123",
    "Nooooo wayyyy 😍",
    "Visit https://test.com"
]
print(full_pipeline(test))
```
